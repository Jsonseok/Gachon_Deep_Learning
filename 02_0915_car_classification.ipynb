```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam
from keras import models,layers,metrics
```

```python
data = pd.read_fwf("/content/car.data", header=None)
data
```

```python
a = []
for i in range(len(data)):
  a.append(data.iloc[i,0].split(','))

train_df = pd.DataFrame(a)
train_df.head()
```

```python
col = ['buying' ,'maint','doors' ,'persons','lug_boot' ,'safety', 'Class'  ]
train_df.columns = col
train_df
```

```python
en = LabelEncoder()
for i in train_df.columns:
  train_df[i] = en.fit_transform(train_df[i])
train_df
```

```python
X = train_df.drop('Class', axis =1)
y = train_df['Class']
```

```python
y_1 = to_categorical(y)
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y_1,test_size=0.2,random_state=0)
```

```python
model = Sequential()

model.add(Dense(6,input_shape=(6,),activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(5,activation='sigmoid'))

model.compile(Adam(learning_rate=0.02),'categorical_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
model_history=model.fit(x=X_train, y=y_train, epochs=100, batch_size=32,validation_data= (X_test,y_test))
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
#Accuracy of the predicted values
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```

```python
a =0
for i,j in zip(y_test_class,y_pred_class):
  if i == j:
    a +=1


print((a/len(y_pred_class))*100)
```
